## 1) Construa um gráfico de barras para avaliar a distribuição das variáveis qualitativas no tempo. Você acha que houve alguma alteração importante a ser investigada?

````python
# importanto as libs
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
````

````python
# carregando e dataset
prev = pd.read_csv('previsao_de_renda.csv')
prev.head()
````

````python
prev_quali = prev[['sexo','posse_de_veiculo', 'posse_de_imovel', 'tipo_renda', 'educacao', 'estado_civil', 
                   'tipo_residencia', 'mau']]
prev_quali
````

````python
# grafico da variavel sexo
ax = sns.countplot(x = 'data_ref', hue ='sexo', data = prev)
tick_labs = prev['data_ref'].unique()


labels = ax.set_xticklabels(tick_labs, rotation = 90)
````

````python
# grafico da variavel posse_de_veiculo
ax = sns.countplot(x = 'data_ref', hue ='posse_de_veiculo', data = prev)
tick_labs = prev['data_ref'].unique()


labels = ax.set_xticklabels(tick_labs, rotation = 90)
````

````python
# grafico da variavel posse_de_imovel
ax = sns.countplot(x = 'data_ref', hue ='posse_de_imovel', data = prev)
tick_labs = prev['data_ref'].unique()


labels = ax.set_xticklabels(tick_labs, rotation = 90)
````

````python
# grafico da variavel tipo_renda
ax = sns.countplot(x = 'data_ref', hue ='tipo_renda', data = prev)
tick_labs = prev['data_ref'].unique()


labels = ax.set_xticklabels(tick_labs, rotation = 90)
````

````python
# grafico da variavel educacao
ax = sns.countplot(x = 'data_ref', hue ='educacao', data = prev)
tick_labs = prev['data_ref'].unique()


labels = ax.set_xticklabels(tick_labs, rotation = 90)
````

````python
# grafico da variavel estado_civil
ax = sns.countplot(x = 'data_ref', hue ='estado_civil', data = prev)
tick_labs = prev['data_ref'].unique()


labels = ax.set_xticklabels(tick_labs, rotation = 90)
````
