#### 1. Baixando dados de ativos
    1. Baixe os dados de três ativos da sua escolha do Yahoo Finance, correspondente a um período de pelo menos 2 anos até a data mais recente.
    2. Salve esses dados para que você possa utilizá-los na próxima tarefa.

```python
# importando pacotes
import yfinance as yf
import pandas as pd
import plotly.express as px
```

```python
# baixar 3 ativos
ativos = ['^BVSP','NVDA', 'BTC-USD']
dt_ini = '2022-01-03'
dt_fim = '2024-01-03'
```

```python
dados = yf.download(ativos, dt_ini, dt_fim)
dados
```

```python
dados.to_csv('ativos.csv', index=False)
```

#### 2. Observando a volatilidade dentro do dia
1. Calcule a amplitude de variação do valor do ativo dentro do dia. Essa amplitude você pode calcular com um comando simples de adicionar uma coluna no Pandas: (valor máximo - valor mínimo).
2. Monte um gráfico de linha mostrando a variação dessa amplitude ao longo do período coletado, por ativo.
3. Veja se consegue identificar que ativo apresentou maior volatilidade.

```python
# criando novo dataframe verticalizado
dados_verticais = dados.stack().reset_index().copy()
dados_verticais
```

```python
# criando coluna da amplitude
dados_verticais['Amplitude'] = dados_verticais['High']-dados_verticais['Low']
dados_verticais
```

```python
# grafico com a variação dos ativos ao longo do tempo
grafico = px.line(dados_verticais, x='Date', y='Amplitude', color='Ticker',title='Variação de BTC, NVDA E BVSP nos últimos 2 anos')

grafico.show()
```

```python
# identificando qual teve maior volatilidade
mais_volatil = dados_verticais.sort_values(by=['Amplitude'],ascending=False).iloc[0]
mais_volatil
```

```python
print(f'O ativo que teve maio volatilidade foi o {mais_volatil['Ticker']}, no dia {mais_volatil['Date']}, onte teve uma variação de {mais_volatil['Amplitude']}')
```
